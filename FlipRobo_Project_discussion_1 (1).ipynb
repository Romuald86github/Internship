Modelling

Normally, for supervised ML, data should be separated into 3 parts: train, test, validate.
but this has a limitation when we use train-test_split module, which separates data only into 2 parts

finding the best random state

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
maxAccu = 0
maxRs = 0
for i in range(1, 200):
    X_train, X_test, y_train, y_test = train_test_split(x1, y1, test_size = 0.30, random_state = i)
    RFR = RandomForestClassifier()
    RFR.fit(X_train,y_train)
    pred = RFR.predict(X_test)
    acc = accuracy_score(y_test, pred)
    if acc>maxAccu:
        maxAccu = acc
        maxRs = i
print('Best accuracy is: ', maxAccu, 'at randome state: ', maxRs)
    
```

creating train test split (using the maxRs above = 100)

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = maxRs)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
"""syntax:
cross_val_score(model_name, features, target variable, cv = foldvalue, scoring = 'accuracy')

in classification algorithms, default scoring is accuracy

in regression, default scoring is r2 score"""
```
