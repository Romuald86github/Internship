```python
driver = webdriver.Chrome()
```

```python
driver.get('https://archive.ics.uci.edu/')
```

```python
url5 = driver.find_element(By.XPATH, '/html/body/div/div[1]/div[2]/ul/li[1]/a')

driver.get(url5.get_attribute('href'))
```

```python
ML_url = []
start = 0
end = 2

for page in range(start, end):
    url = driver.find_elements(By.XPATH, '//a[@class="link-hover link text-xl font-semibold"]')
    for i in url:
        ML_url.append(i.get_attribute('href'))
    button = driver.find_element(By.XPATH, '//button[@aria-label="Next Page"]')
    button.click()
    time.sleep(2)
```

```python
dataset_name =[]
datatype = []
task = []
attribute_type = []
N_instances = []
N_attribute = []
year = []
```

```python
for url in ML_url:
    driver.get(url)
    time.sleep(2)
    
    try:
        d_n = driver.find_elements(By.XPATH, '//h1[@class="text-3xl font-semibold text-primary-content"]')
        for i in d_n:
            dataset_name.append(i.text)
            
    except NoSuchElementException:
        dataset_name.append('_')
        
    
    try:
        d_t = driver.find_elements(By.XPATH, '//div[1][@class="col-span-4"]')
        for i in d_t:
            datatype.append(i.text)
            
    except NoSuchElementException:
        datatype.append('_')
        
        
    try:
        t = driver.find_elements(By.XPATH, '//div[3][@class="col-span-4"]')
        for i in t:
            task.append(i.text)
            
    except NoSuchElementException:
        task.append('_')
        
        
    try:
        attr = driver.find_elements(By.XPATH, '//div[4][@class="col-span-4"]')
        for i in attr:
            attribute_type.append(i.text)
            
    except NoSuchElementException:
        attribute_type.append('_')
        
        
        
    try:
        NI = driver.find_elements(By.XPATH, '//div[5][@class="col-span-4"]')
        for i in NI:
            N_instances.append(i.text)
            
    except NoSuchElementException:
        N_instances.append('_')
        
    
    try:
        NA = driver.find_elements(By.XPATH, '//div[6][@class="col-span-4"]')
        for i in NA:
            N_attribute.append(i.text)
            
    except NoSuchElementException:
        N_attribute.append('_')
        
        
        
    try:
        ye = driver.find_elements(By.XPATH, '//h2[@class="text-sm text-primary-content"]')
        for i in ye:
            year.append(i.text)
            
    except NoSuchElementException:
        year.append('_')
        
```

```python
print(len(dataset_name), len(datatype), len(task), len(attribute_type), len(N_instances), len(N_attribute), len(year))
```

```python
df5 = pd.DataFrame({'dataset_name':dataset_name[0:19], 'datatype':datatype[0:19], 
                   'task':task[0:19], 'attribute_type':attribute_type[0:19], 'N_instances':N_instances[0:19], 'N_attribute':N_attribute[0:19], 'year':year})
df5
```
